```python
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
```

```python
count, bins, patches = plt.hist(s, 30, normed=True) #hist
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * 
         np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r') #pdf
plt.show()
```
