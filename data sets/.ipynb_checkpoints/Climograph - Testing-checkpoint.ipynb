```python
fig, ax1 = plt.subplots()
t = np.arange(0.01, 10.0, 0.01)
s1 = np.exp(t)
plt.plot(t,s1,'b-')
plt.xlabel('t (s)')
plt.ylabel('exp',color='b')
 
ax2 = ax1.twinx()
s2 = np.sin(2*np.pi*t)
ax2.plot(t, s2, 'r.')
plt.ylabel('sin', color='r')
plt.show()
```

```python
###### SJ dataset's climograph
#laying the first axes 
x = list(range(12))
xticks = list(sj)
fig, ax1 = plt.subplots(figsize=(6, 8))
#first axes contains the precipitation
s1 = sj.iloc[2]
plt.bar(x, s1, width=0.5, color="blue", align="center")
plt.xlabel('Months')
plt.ylabel('Precipitation (in.)', color='b')
plt.xticks(x,xticks)
plt.xlim(min(x),max(x))
plt.ylim([0,20])

#laying the second axes, contains the temperature
ax2 = ax1.twinx() #review
s2 = sj.iloc[0] 
s3 = sj.iloc[1]
plt.plot(x,s2, 'r')
plt.plot(x,s3, 'r')
plt.ylabel('Temperature (F)', color = 'r')
plt.xlim(min(x)-0.5,max(x)+0.5)
plt.ylim([-40,120])

#drawing 0 degrees line
plt.plot(x, np.zeros_like(x), 'k--')

plt.show()
```
